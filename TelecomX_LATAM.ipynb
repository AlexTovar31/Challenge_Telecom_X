```python
import pandas as pd
import requests
import json
```

```python
url = ('https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json')
```

```python
response = requests.get(url)
data = response.json()
telecom = pd.DataFrame(data)
```

```python
telecom = pd.json_normalize(data)
telecom.rename(columns={'account.Charges.Monthly': 'Charges_Monthly',
                        'account.Charges.Total': 'Charges_Total'}, inplace=True)
telecom.rename(columns=lambda x: x.split('.')[-1], inplace=True)
telecom.sample(11)
```

```python
telecom.info()
```

#ðŸ”§ TransformaciÃ³n

```python
import numpy as np
```

##Eliminacion de caracteres especiales y transformacion a tipo float

```python
telecom['Charges_Total'] = telecom['Charges_Total'].str.replace(',', '').replace('$', '').replace(' ', '').replace('', None)
#Tranformo a float
telecom['Charges_Total'] = telecom['Charges_Total'].astype(np.float64)
#reviso
telecom['Charges_Total'].info()
```

```python
#identifico por quÃ© estÃ¡n vacÃ­as esas filas.
Charges_nulos = telecom['Charges_Total'].isna()
telecom[Charges_nulos]
```

##Eliminacion de los datos nulos de la columna Charges_total

```python
telecom = telecom.drop(telecom[Charges_nulos].index, axis=0)
telecom = telecom.reset_index(drop=True)
```

```python
telecom.head()
```

##Remplazamos los valores vacios por el No en la columna Churn por que tiene informacion con otras colunas para el analisis.

```python
churn_nulos = telecom['Churn'].str.strip() == ''

telecom[churn_nulos]
```

```python
telecom['Churn'] = telecom['Churn'].replace('', 'No')
```

##Grafica Porcentaje de cliente de permanencia y cancelacion.

```python
fig, ax = plt.subplots(figsize=(8, 6))

churn_counts = telecom['Churn'].value_counts()
churn_porcentage = telecom['Churn'].value_counts(normalize=True) * 100

bars = ax.bar(churn_counts.index, churn_counts.values, color =['skyblue', 'red'])
ax.set_title('Porcentaje de clientes de permanencia y cancelacion', loc ='left', fontsize = 18)
ax.set_xlabel('Churn')
ax.set_ylabel('Cantidad de clientes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.xticks(['No', 'Yes'], ['Permanecen', 'Abandonaron'])

labels = [f'{p:.1f}%' for p in churn_porcentage[churn_counts.index]]

ax.bar_label(bars, labels=labels, padding=3, fontsize=12)

ax.grid(linestyle='--')
plt.tight_layout()

fig.savefig('churn_porcentage.png')

plt.show()
```

##Grafica Comparacion de clientes que abandonaron por tipo de contrato



```python
# Agrupar por tipo de contrato y Churn, y contar ocurrencias
contract_churn = telecom.groupby(['Contract', 'Churn']).size().unstack(fill_value=0)

ax = contract_churn.plot(kind='bar', figsize=(10, 6), color=['blue', 'red'])

# Etiquetas y tÃ­tulo
ax.set_title('ComparaciÃ³n de clientes que abandonaron por Tipo de Contrato', loc='left', fontsize=18)
ax.set_xlabel('Tipo de Contrato', fontsize=14)
ax.set_ylabel('Cantidad de Clientes', fontsize=14)
ax.tick_params(axis='x', rotation=0) # Rotar etiquetas del eje x si es necesario
ax.legend(title='Churn', labels=['Permanecen', 'Abandonaron'])

# Mostrar el porcentaje en las barras
for container in ax.containers:
    ax.bar_label(container, fmt='%d')

ax.grid(axis='y', linestyle='--')
plt.tight_layout()

fig.savefig('comparacion_churn_contrato.png')

plt.show()
```

##'DistribuciÃ³n de Clientes que Permanecen por Tipo de Contrato y Meses de permanencia.

```python
# prompt: DistribuciÃ³n de Clientes que Permanecen por Tipo de Contrato y tenure.

# Filter for customers who stayed (Churn == 'No')
stayed_customers = telecom[telecom['Churn'] == 'No'].copy()

# Create bins for tenure
stayed_customers['tenure_bins'] = pd.cut(stayed_customers['tenure'],
                                         bins=[0, 6, 12, 24, 48, 60, 72],
                                         labels=['0-6', '7-12', '13-24', '25-48', '49-60', '61-72'],
                                         right=False)

# Group by Contract and tenure_bins and count the number of customers
stayed_distribution = stayed_customers.groupby(['Contract', 'tenure_bins']).size().unstack(fill_value=0)

# Plotting
ax = stayed_distribution.plot(kind='bar', figsize=(14, 7), cmap='viridis')

# Labels and title
ax.set_title('DistribuciÃ³n de Clientes que Permanecen por Tipo de Contrato y Meses de permanencia', loc='left', fontsize=18)
ax.set_xlabel('Tipo de Contrato', fontsize=14)
ax.set_ylabel('Cantidad de Clientes', fontsize=14)
ax.tick_params(axis='x', rotation=0)
ax.legend(title='Meses de permanencia')

# Show the count on the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d')

ax.grid(axis='y', linestyle='--')
plt.tight_layout()

plt.show()
```

##Grafica de Metodos de pagos por clientes que abandonaron

```python
churned_customers = telecom[telecom['Churn'] == 'Yes']

# Contar la distribuciÃ³n de PaymentMethod para estos clientes
payment_method_counts_churned = churned_customers['PaymentMethod'].value_counts()

# Crear el grÃ¡fico de barras
fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.bar(payment_method_counts_churned.index, payment_method_counts_churned.values, color=sns.color_palette("viridis", len(payment_method_counts_churned)))

ax.set_title('MÃ©todos de Pago para Clientes que Abandonaron', loc='left', fontsize=18)
ax.set_xlabel('MÃ©todo de Pago', fontsize=14)
ax.set_ylabel('Cantidad de Clientes', fontsize=14)
ax.tick_params(axis='x', rotation=45, labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# Agregar las etiquetas de valor en las barras
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center', fontsize=10) # va: vertical alignment, ha: horizontal alignment

ax.grid(axis='y', linestyle='--')
plt.tight_layout()

fig.savefig('payment_method_churned.png')
plt.show()
```

```python
# prompt: crear grafica de barras de churn que abandonaron con la columna Charges_Monthly

# Filta los clientes que abandonaron
churned_customers = telecom[telecom['Churn'] == 'Yes'].copy()

# Crea bins para Charges_Monthly
churned_customers['Charges_Monthly_bins'] = pd.cut(churned_customers['Charges_Monthly'],
                                                  bins=[0, 20, 40, 60, 80, 100, 120],
                                                  labels=['0-20', '21-40', '41-60', '61-80', '81-100', '101-120'],
                                                  right=False)

# Contar la distribuciÃ³n de Charges_Monthly_bins para estos clientes
monthly_charges_distribution_churned = churned_customers['Charges_Monthly_bins'].value_counts().sort_index()

# Crear el grÃ¡fico de barras
fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.bar(monthly_charges_distribution_churned.index, monthly_charges_distribution_churned.values, color=sns.color_palette("magma", len(monthly_charges_distribution_churned)))

ax.set_title('DistribuciÃ³n de Cargos Mensuales para Clientes que Abandonaron', loc='left', fontsize=18)
ax.set_xlabel('Cargos Mensuales', fontsize=14)
ax.set_ylabel('Cantidad de Clientes', fontsize=14)
ax.tick_params(axis='x', rotation=45, labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# Agregar las etiquetas de valor en las barras
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center', fontsize=10) # va: vertical alignment, ha: horizontal alignment

ax.grid(axis='y', linestyle='--')
plt.tight_layout()

fig.savefig('monthly_charges_churned.png')
plt.show()
```

```python
# ##Grafica Comparativa entre columna Churn y columna InternetService
internet_churn = telecom.groupby(['InternetService', 'Churn']).size().unstack(fill_value=0)

ax = internet_churn.plot(kind='bar', figsize=(10, 6), color=['green', 'red'])

ax.set_title('Comparativa de clientes que abandonaron por Servicio de Internet', loc='left', fontsize=18)
ax.set_xlabel('Servicio de Internet', fontsize=14)
ax.set_ylabel('Cantidad de Clientes', fontsize=14)
ax.tick_params(axis='x', rotation=0)
ax.legend(title='Churn', labels=['Permanecen', 'Abandonaron'])

for container in ax.containers:
    ax.bar_label(container, fmt='%d')

ax.grid(axis='y', linestyle='--')
plt.tight_layout()

fig.savefig('internet_churn.png')
plt.show()
```
